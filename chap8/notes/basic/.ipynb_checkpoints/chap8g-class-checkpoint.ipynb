## 1. 什么是静态变量

我们在C++中都学过静态变量，用static进行修饰的变量就是静态变量

```python
import objgraph
```

```python
class Number():
    num = 0

    @staticmethod
    def printNum():
        print(Number.num)
        
    def printNum(this):
        print(this.num)
```

```python
objgraph.show_refs([Number], filename="Number_Class.png")
```

```python
n1 = Number()
n2 = Number()
print(n1.num, n2.num, Number.num)
objgraph.show_refs([n1.num,n2.num, Number.num], filename="static_member.png")
```

```python
n1.num += 2
print(n1.num, n2.num, Number.num)
objgraph.show_refs([n1.num,n2.num, Number.num], filename="static_member.png")
```

```python
n2.num += 3
print(n1.num, n2.num, Number.num)
print(id(n1), id(n2), id(Number))
objgraph.show_refs([n1,n2, Number], filename="static_member.png", max_depth=2)
```

```python
class TaskQueue():
    queue=[]
    
    def append(self,obj):
        self.queue.append(obj)
    
    def print_queue(self):
        print(self.queue)
```

```python
queue1 = TaskQueue()
queue2 = TaskQueue() 

objgraph.show_refs([queue1, queue2])
```

```python
queue1.append(1)
objgraph.show_refs([queue1, queue2])
```

```python
queue2.append(2)
objgraph.show_refs([queue1.queue, queue2.queue])
```

```python
print(TaskQueue.queue)
```

从以上现象我们可以得到什么结论？

如果静态变量为可变对象Mutable object，则有什么现象？

如果静态变量为不可变对象Immutable object，又会发生什么？

- A tuple with only one element will not be treated as a tuple. However, it will be treated as an atomic type - a number, a string, a boolean, or None.

```python
t = (2.0)
type(t)
```

But we can convert it into a tuple in this fashion (using a comma behind the only element):

```python
t = (2.0,)
type(t)
```

What about a list? Is it similar or different?

```python
li = [1]
type(li)
```

#### list/tuple/dict/str都是Iterable object?

```python
from collections.abc import Iterable
print(isinstance(list(), Iterable))
print(isinstance(dict(), Iterable))
print(isinstance(tuple(), Iterable))
print(isinstance("", Iterable))
```

也可以说，是包含了`__iter__()`方法的对象。

```python
print("__iter__" in dir(list()))
print("__iter__" in dir(dict()))
print("__iter__" in dir(tuple()))
print("__iter__" in dir(""))
```

#### 但它们都不是Iterator

> 记住，根据Iterator协议，一个Iterator必须包含`__next__()`方法。

```python
print("__next__" in dir(list()))
print("__next__" in dir(dict()))
print("__next__" in dir(tuple()))
print("__next__" in dir(""))
```
